```
meds = ['metformin', 'repaglinide', 'nateglinide', 'chlorpropamide',
       'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide',
       'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone',
       'tolazamide', 'insulin', 'glyburide-metformin', 'glipizide-metformin',
       'glimepiride-pioglitazone', 'metformin-rosiglitazone',
       'metformin-pioglitazone']
```

```
remove = []
remove_2 = []
for i in meds:
    print(i, '\n', df[i].value_counts(), '\n')
    vc = df[i].value_counts().values
    if len(vc) > 1:
        if(vc[1] < 25):
            remove.append(i)
    else:
        remove.append(i)
    if vc[1:].sum() < 0.01 * vc.sum():
        remove_2.append(i)
```

```
remove_2, len(remove_2)
```

```
insigni_meds = ['nateglinide', 'chlorpropamide', 'acetohexamide', 'glyburide', 'tolbutamide', 'rosiglitazone', 'acarbose',
 'miglitol', 'troglitazone', 'tolazamide', 'glyburide-metformin', 'glipizide-metformin', 'glimepiride-pioglitazone',
 'metformin-rosiglitazone', 'metformin-pioglitazone']
```

```
remove_meds = []
for i in meds:
    if i in remove_2 and i in insigni_meds:
        remove_meds.append(i)  
```

```
X = test_df.drop('readmitted', axis=1)
y = test_df['readmitted']

X_dum = pd.get_dummies(X, drop_first = True)


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_dum, y, test_size = 0.3,random_state = 3, stratify = df['readmitted'])


from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=pd.DataFrame(ss.fit_transform(X_train),columns=X_train.columns)
X_test=pd.DataFrame(ss.transform(X_test),columns=X_test.columns)

from imblearn.over_sampling import SMOTE 
sm = SMOTE(random_state = 2) 
X_train_sm, y_train_sm = sm.fit_sample(X_train, y_train) 

import time
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

RFC=RandomForestClassifier()


### Without smote
print('Without smote')

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

SKF=StratifiedKFold(n_splits=5,shuffle=True, random_state = 5)

cv_scores={}
for score in ["f1", "f1_weighted", "roc_auc", "recall"]:
    cv_results = cross_val_score(RFC, X_train, y_train, cv=SKF, scoring=score,n_jobs=-1)
    cv_scores[score] ="%f (%f)" % (cv_results.mean(), cv_results.std())
    #print(cv_results)
print(cv_scores)


### With SMOTE
print()
print('Smote')
cv_scores_sm={}
for score in ["f1", "f1_weighted", "roc_auc", "recall"]:
    cv_results_sm = cross_val_score(RFC, X_train_sm, y_train_sm, cv=SKF, scoring=score,n_jobs=-1)
    cv_scores_sm[score] ="%f (%f)" % (cv_results_sm.mean(), cv_results_sm.std())
    #print(cv_results_sm)
print(cv_scores_sm)
```

```
X = df.drop('readmitted', axis=1)
y = df['readmitted']

X_dum = pd.get_dummies(X, drop_first = True)


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_dum, y, test_size = 0.3,random_state = 3, stratify = df['readmitted'])

from imblearn.over_sampling import SMOTE 
sm1 = SMOTE(random_state = 3) 
X_train_sm, y_train_sm = sm1.fit_sample(X_train, y_train) 

import time
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

RFC=RandomForestClassifier(random_state = 3, n_jobs = -1)

RFC.fit(X_train, y_train)
pred = RFC.predict(X_test)
```

```
from sklearn.metrics import classification_report
```

```
print(classification_report(y_test, pred))
```

```
RFC=RandomForestClassifier(random_state = 3, n_jobs = -1)

RFC.fit(X_train_sm, y_train_sm)
pred = RFC.predict(X_test)
print(classification_report(y_test, pred))
```

```
sm2 = SMOTE(random_state = 3, sampling_strategy='not minority')
X_train_sm, y_train_sm = sm2.fit_sample(X_train, y_train)
RFC=RandomForestClassifier(random_state = 3, n_jobs = -1)

RFC.fit(X_train_sm, y_train_sm)
pred = RFC.predict(X_test)
print(classification_report(y_test, pred))
```
