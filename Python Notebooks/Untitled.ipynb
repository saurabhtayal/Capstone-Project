```python
diag_cols = ['diag_1','diag_2','diag_3']
```

```python
for col in diag_cols:
    df[col] = df[col].str.replace('E','-')
    df[col] = df[col].str.replace('V','-')
    df[col] = df[col].str.replace('?','0')
    if '250' in df[col]:
        df.loc[col] = '250'
```

```python
df[diag_cols] = df[diag_cols].astype(float)
```

```python
for col in diag_cols:
    df['temp']=np.nan
    
    condition = df[col]==250
    df.loc[condition,'temp']='Diabetes'
    
    condition = (df[col]>=390) & (df[col]<=458) | (df[col]==785)
    df.loc[condition,'temp']='Circulatory'
    
    condition = (df[col]>=460) & (df[col]<=519) | (df[col]==786)
    df.loc[condition,'temp']='Respiratory'
    
    condition = (df[col]>=520) & (df[col]<=579) | (df[col]==787)
    df.loc[condition,'temp']='Digestive'
    
    condition = (df[col]>=580) & (df[col]<=629) | (df[col]==788)
    df.loc[condition,'temp']='Genitourinary'
    
    condition = (df[col]>=800) & (df[col]<=999)
    df.loc[condition,'temp']='Injury'
    
    condition = (df[col]>=710) & (df[col]<=739)
    df.loc[condition,'temp']='Muscoloskeletal'
    
    condition = (df[col]>=1) & (df[col]<=139)
    df.loc[condition,'temp']='Parasites'
    
    condition = ((df[col]>=790) & (df[col]<=799)) | (df[col] == 780) | (df[col] == 781) | (df[col] == 784)
    df.loc[condition,'temp']='Ill-defined conditions'
    
    condition = (df[col]>=240) & (df[col]<=279) & (df[col]!=250)
    df.loc[condition,'temp']='Metabolic diseases and immunity disorders'
    
    condition = (df[col]>=680) & (df[col]<=709) & (df[col]==782)
    df.loc[condition,'temp']='Diseases of the skin and subcutaneous tissue'
    
    condition = (df[col]>=290) & (df[col]<=319)
    df.loc[condition,'temp']='Mental disorders'
    
    condition = (df[col]>=280) & (df[col]<=289)
    df.loc[condition,'temp']='Diseases of the blood and blood-forming organs'
    
    condition = (df[col]>=320) & (df[col]<=359)
    df.loc[condition,'temp']='Diseases of the nervous system'
    
    condition = (df[col]>=630) & (df[col]<=679)
    df.loc[condition,'temp']='Complications of pregnancy, childbirth, and the puerperium'
    
    condition = (df[col]>=360) & (df[col]<=389)
    df.loc[condition,'temp']='Diseases of the sense organs'
    
    condition = (df[col]>=740) & (df[col]<=759)
    df.loc[condition,'temp']='Congenital anomalies'
    
    condition = df[col]==0
    df['temp']=df['temp'].fillna('Others')
    df[col]=df['temp']
    df.drop('temp',axis=1,inplace=True)
```
